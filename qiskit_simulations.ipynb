```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from qiskit.quantum_info import Statevector, random_statevector
from qiskit import QuantumCircuit, Aer, execute
import qiskit.visualization as qvi

import qiskit.tools.jupyter
%qiskit_version_table
%matplotlib inline
```

Pre-loading states $\left|0\right\rangle$ and $\left|1\right\rangle$.

```python
# States |0> and |1>
state_0 = Statevector.from_label('0')
state_1 = Statevector.from_label('1')
```

### Exercise 1
1-qubit quantum circuit outputting a superposition state $\left.\left.\frac{1}{\sqrt{2}} \right(\left|0\right\rangle + 
\left|1\right\rangle\right)$ when the initial state is $\left|0\right\rangle$ .

```python
# Initial state vector
init_vec = state_0

#Creat a 1-bit quantum circuit
circuit = QuantumCircuit(1) 

# Set initial state to zero
circuit.initialize(init_vec, 0)

# Add Hadamard gate to the circuit
circuit.h(0)

# Vizualize the circuit
circuit.draw(output='mpl')
```

```python
# Simulate
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
result = job.result()

# Get the final statevector
final_vec = result.get_statevector()
```

Vizualizing the initial and final states in the Bloch sphere.

```python
qvi.plot_bloch_multivector(init_vec, title= 'Initial state')
```

```python
qvi.plot_bloch_multivector(final_vec, title= 'Final state')
```

### Exercise 5
2-qubit quantum circuit for generating a Bell state $\beta_{00}$

```python
# Define the quantum circuit with 2 qubits
circuit = QuantumCircuit(2)

# Set initial state to zero
circuit.initialize(state_0, [0]) # First qubit
circuit.initialize(state_0, [1]) # Second qubit

# Initial state measure
init_vec = Statevector(circuit)

## Implement the Bell state generator circuit

# Apply the Hadamard gate to the first qubit of the circuit
circuit.h(0)

# Add the CNOT gate 
circuit.cx(0,1) # (control, target)

# Vizualize the circuit
circuit.draw(output='mpl')
```

```python
# Simulate
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
result = job.result()

# Get the final statevector
final_vec = result.get_statevector()

# Visualize the final state
final_vec.draw('latex')
```

Vizualizing the initial states in the Bloch Sphere. It is possible to do so since the initial state vectors are not entangled.

```python
# Visualizing the initial state in a bloch sphere
qvi.plot_bloch_multivector(init_vec, title= 'Initial state')
```

For the final state, it is not possible to visualize them in a Bloch Sphere because they are entangled states. In this case, the "qsphere" representation was used.

```python
qvi.plot_state_qsphere(final_vec, show_state_labels=True, show_state_phases=True, use_degrees=False)
```

### Exercise 7
Circuit including a 3-qubit quantum register and a 3-bit 
classical register so that the initial state is $\left|010\right\rangle$ and the final state is $\left|001\right\rangle$.

```python
# Define the quantum circuit with 3 qubits and 3 classical bits
circuit = QuantumCircuit(3,3)

# Set initial state to zero
circuit.initialize(state_0, [0]) # First qubit
circuit.initialize(state_1, [1]) # Second quibit
circuit.initialize(state_0, [2]) # Second quibit

# Initial state measure
init_vec = Statevector(circuit)

circuit.cnot(1, 0) 
circuit.cnot(0, 1)

#circuit.barrier(range(3))

# Measure q_1 and q_2
circuit.measure([1, 2], [1, 2])

# Visualizing the circuit
circuit.draw(output='mpl')
```

Simulating the circuit.

```python
# Simulate
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
result = job.result()

# Get the final statevector
final_vec = result.get_statevector()

# Visualize the final state
final_vec.draw('latex')
```

Visualizing the initial state in the Bloch sphere and the final state, **which is a pure state**, using the "qsphere" representation.

```python
# Visualizing the initial state in a bloch sphere
qvi.plot_bloch_multivector(init_vec, title= 'Initial state')
```

```python
# Visualizing the final state 
qvi.plot_state_qsphere(final_vec)
```

Visualizing the counts histogram. The histogram result are not coherent with the output state obtained previously. This happens because we are not measuring the first qubit.

```python
# Returns counts
qvi.plot_histogram(result.get_counts())
```

### Exercise 8
Generating a random state vector.

```python
# Generate a random quantum state
init_psi = random_statevector(2)

# Vizualize the state in on a Bloch sphere
qvi.plot_bloch_multivector(init_psi, title= 'Initial random state')
```

Programming the quantum circuit with the barrier and the measurement and simulating it.

```python
# Define the quantum circuit with 1 qubits and 1 classical bit
circuit = QuantumCircuit(1,1)

# Set initial state to psei
circuit.initialize(init_psi, [0]) 

# Add a barrier
circuit.barrier()

# Add the measurement
circuit.measure(0,0) 

# Vizualize the circuit
circuit.draw(output='mpl')
```

```python
# Simulate
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
result = job.result()

# Get the final statevector
final_psi = result.get_statevector()
```

Visualizing the output state vector and the counts histogram.

```python
# Visualizing the final state on a bloch sphere
qvi.plot_bloch_multivector(final_psi, title= 'Final state')
```

```python
# Returns counts
qvi.plot_histogram(result.get_counts())
```

The input state vector is random $\alpha_1 \left|0\right\rangle + \alpha_2\left|1\right\rangle$ vector and collapses into either the state $\left|0\right\rangle$ or $\left|1\right\rangle$ after measurement with probability given by $\alpha_i^2$ that can be visualized by the proximity to the corresponding pole on the Bloch sphere. Each time re reinitialize the random state, we obtain a different count that dependes on the coefficients $\alpha_i$.

### Exercise 9
Programming the proposed quantum teleportation circuit.

```python
# Define the quantum registers and classical registers separately
quantum_register = qiskit.QuantumRegister(3, name='q')
crz, crx = qiskit.ClassicalRegister(1, name='crz'), qiskit.ClassicalRegister(1, name='crx')
circuit = QuantumCircuit(quantum_register, crz, crx)

# Initialize the circuit with a random state vector for q0
init_psi = random_statevector(2)
circuit.initialize(init_psi, [0]) # q0
circuit.initialize(state_0,  [1]) # q1
circuit.initialize(state_0,  [2]) # q2

# Add the gates to the circuit
circuit.h(1)
circuit.cx(1,2)
circuit.barrier(range(3))
circuit.cx(0,1)
circuit.h(0)
circuit.barrier(range(3))
circuit.measure(0, crz)
circuit.measure(1, crx)

# Visualize the circuit
circuit.draw(output='mpl')
```

```python
# Add conditional gates
circuit.x(2).c_if(crx, 1)
circuit.z(2).c_if(crz, 1)

# Visualize the circuit
circuit.draw(output='mpl')
```

Simulating the circuit and then visualizing the initial and final states on the Bloch sphere.

```python
# Simulate
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
result = job.result()

# Get the final statevector
final_psi = result.get_statevector()
```

```python
# Visualizing the initial state 
qvi.plot_bloch_multivector(init_psi, title= 'Initial state')
```

```python
# Visualizing the final state 
qvi.plot_bloch_multivector(final_psi, title= 'Final state')
```

It is possible to notice that the initial state vector is re-obtained as an output for the circuit. Meaning that teleportation is achieved.
